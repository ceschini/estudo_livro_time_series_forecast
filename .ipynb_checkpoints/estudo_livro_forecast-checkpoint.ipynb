### Capítulo 2: O que é Previsão de Séries Temporais?

Previsão de séries temporais é uma área importante de machine learning que é normalmente neglicenciada devido sua complexidade. Sua importância se dá devido ao grande número de problemas de previsão que envolvem tempo.

#### Séries Temporais

A coleção de dados de uma aplicação de machine learning normalmente é um conjunto de observações, esses conjuntos normalmente possuem um fator de tempo que é usado para prever resultados de novos dados. No caso de dados de séries temporais, é adicionado uma dependencia de ordem entre os dados, a dimensão de tempo. Essa dimensão é uma restrição e também uma estrutura que permite diferentes coletas de informações.

##### conjunto comum de dados de Machine Learning (coleção de observações):

observation #1

observation #2

observation #3

#### conjunto de dados de Séries Temporais (sequência de observações pegas sequencialmente no tempo): 

Time #1, observation

Time #2, observation

Time #3, observation

#### Nomenclatura

É preciso definir os termos padrões dos dados de séries temporais. O tempo atual é definido como **t**, uma observação nesse tempo atual é definido como **obs(t)**.

As vezes estamos interessados em observações feitas em um tempo prévio, chamadas de *lags*. Tempos passados são negativos em relação ao tempo atual, por exemplo, o tempo anterior é **t-1** e o tempo antes desse é **t-2**. Observações nesses tempos são **obs(t-1)** e **obs(t-2)** respectivamente. Tempos no futuro é o que estamos interessados em prever e são positivos em relação ao tempo atual. Por exemplo o próximo tempo é **t+1** e o tempo depois desse é **t+2**. As observações nesses tempos são **obs(t+1)** e **obs(t+2)** respectivamente.

Por simplicidade, as vezes é deixado de lado a notação **obs(t)** e usamos **t+1** no lugar, assume-se que quando estamos falando de um tempo, é sobre a observação dele e não o seu índice. Adicionalmente ao se refer a algum lag, pode ser usado a notação *lag de 10* ou **lag=10** ao invés de **t-10**. Resumindo:

- t-n: um tempo passado, lag (ex: t-1 para o tempo prévio)
- t: tempo atual e ponto de referência
- t+n: um tempo futuro ou tempo de previsão (ex: t+1 para o próximo tempo)

#### Descrevendo vs. Prevendo

Existe dois objetivos quando se trata de séries temporais, descrição e previsão. Descrição é a análise das séries temporais, onde os dados são estudados e modelados para ter um melhor entendimento e assim auxiliar na predição dos dados. Porém, não está diretamente ligado a previsão.

Na modelagem descritiva, ou análise de séries temporais, as séries temporais são modeladas para identificar padrões de temporadas, tendências, relações com fatores externos, etc. Já em previsões de séries temporais, o objetivo é prever os valores das próximas séries.

##### Análise de Séries Temporais

Ao aplicar estatística clássica nas séries temporais, o principal foco é a análise. O objetivo de se analisar as séries temporais é criar modelos que expliquem qual o problema em questão, determinando suas características e suas causas. Essa área de estudo busca o *por que* atrás das séries temporais. A qualidade do modelo se dá na sua capacidade de modelar todos os dados e informar o problema deles.

##### Previsão de Séries Temporais

O principal objetivo é, a partir de um modelo descritivo, prever os próximos valores em uma série de dados temporais. A qualidade do estudo se dá na sua eficácia de prever o futuro. Um dado importante é a falta de acesso aos dados futuros, sendo possível somente estimá-lo a partir dos dados disponíveis.

#### Componentes de Séries Temporais

A análise de séries temporais providencia diversas ferramentas para entender melhor os conjuntos. Um dos melhores desses é a decomposição das séries em 4 partes:

* **Nível**. O valor base para a série se ela fosse uma linha reta.
* **Tendência**. O comportamento de aumento ou decrimento normalmente linear de uma série ao longo do tempo. (Opcional)
* **Sazonalidade**. O comportamento de padrões repetitivos ou ciclos ao longo do tempo. (Opcional)
* **Ruído**. Variabilidade nas observações que não podem ser explicados pelo modelo. (Opcional)

Todas séries temporais tem um nível, a maioria tem ruído, e a tendência e sazonalidade são opcionais.

A principal funcionalide de diversas séries temporais são as variações sazonais e as tendências. Outra funcionalidade importante na maioria das séries temporais é que observações próximas no tempo tendem a estar correlacionadas (dependentes serialmente).

#### Interesses da Previsão

Séries temporais frequentemente requerem limpezas, escalas e até transformações. Por exemplo:

* **Frequencia**. Talvez os dados são dispostos em uma frequência muito alta ou os valores estão desnivelados de uma forma que não seja possível modelá-los, nesses casos é necessário reorganizar as amostras.
* **Pontos fora da curva**. Talves tenha pontos corruptos ou fora da curva que precisa serem investigados e lidados.
* **Faltando**. Talvez tenha *gaps* ou dados faltando que precisa de interpolação ou imputação.

Normalmente os problemas de séries temporais são *real-time*, continualmente providenciando novas oportunidades de previsões. Isso adiciona uma honestidade à previsão que rapidamente aponta possíveis erros de modelagem ou suposições ruins.

#### Exemplos de Previsões de Séries Temporais

Segue exemplos de uma variedade de indústrias para fazer as noções de análise e previsão de séries temporais mais concretas.

* Previsão do rendimento do milho em toneladas por estado a cada ano.
* Prever se um traçado de EEG em segundos indica que um paciente está tendo uma convulsão ou não.
* Previsão do preço de fechamento de uma ação a cada dia.
* Previsão da taxa de natalidade em todos os hospitais de uma cidade a cada ano.
* Previsão de vendas de produtos em unidades vendidas a cada dia para uma loja.
* Previsão de desemprego para um estado a cada trimestre.
* Previsão do preço médio da gasolina em uma cidade a cada dia.
* Previsão da demanda de utilização em um servidor a cada hora.

### Capítulo 3: Séries Temporais como *Supervised Learning*

#### Supervised Machine Learning

A maioria das práticas em *machine learning* utilizam *supervised learning*. *Supervised Learning* é quando você tem variáveis de entrada (X) e uma variável de saída (y) e você usa um algoritmo para aprender a função de mapeamento da entrada para a saída.

  #####                                                *Y = f(X)

O objetivo é aproximar o mapeamento real de forma quase perfeita, para quando você tiver novos dados de entrada (X), você puder prever as variáveis de saída (Y) para esses dados. Abaixo está um exemplo de um *dataset* de *supervised learning* onde cada linha é uma observação compreendida de uma variável de entrada (X) e uma variável de saída para ser prevista (y).

X, y

5, 0.9

4, 0.8

5, 1.0

3, 0.7

4, 0.9

Se diz que essa abordagem é supervisionada (supervised) pois o processo de ensinar um algoritmo a partir do dataset de treinamento pode ser pensado como um professor supervisionando o processo de aprendizagem. Nós sabemos as respostas corretas; o algoritmo iterativamente faz predições nos dados de treinamento e é corrigido fazendo atualizações. O aprendizado para quando o algoritmo alcança um nível aceitável de performance. Os problemas dessa abordagem podem futuramente ser agrupados em problemas de regressão e classificação.

- **Classificação**: Um problema de classificação é quando a variável de saída é uma categoria, como **vermelho** e **azul** ou **doença** e **sem doença**.
- **Regressão**: Um problema de regressão é quando a variável de saída é um valor real, como **dollars** ou **peso**. O exemplo acima é um problema de regressão.

#### Janela Deslizante

Dados de séries temporais podem ser expressos como *supervised learning*. Dado uma sequência de números para um *dataset* de séries temporais, nós podemos reestruturar os dados para parecer um problema de aprendizagem supervisionada. Nós podemos fazer isso ao utilizar passos temporais prévios como variáveis de entrada e usar os próximos passos temporais como a variável de saída. Vamos concretizar isso com um exemplo. Imagine que temos uma série temporal como segue:

tempo, medida

1, 100

2, 110

3, 108

4, 115

5, 120

Nós podemos reestruturar esse dataset como um problema de aprendizagem supervisionada utilizando os valores de um passo anterior para prever os valores do próximo passo. Reorganizando o dataset dessa forma, os dados ficariam como segue:

X, y

?, 100

100, 110

110, 108

108, 115

115, 120

120, ?

Dê uma olhada no dataset transformado acima e compare com o original. Aqui estão algumas observações:

- Nós podemos ver que o passo anterior é a entrada (X) e o próximo passo é a saída (Y) no nosso problema de aprendizado supervisionado
- Nós podemos ver que a ordem entre as observações foi preservada, e deve ser mantida no futuro para treinar o modelo supervisionado
- Nós podemos ver que não temos valor prévio que pode ser usado para prever o primeiro valor na sequência. Nós iremos deletar essa linha já que não podemos utilizá-la.
- Nós também podemos ver que não temos um próximo valor conhecido para prever para o último valor na sequência. Podemos deletar esse valor enquanto treinamos nosso modelo supervisionado.

O uso de passos prévios para prever os próximos é chamado de método de janela deslizante. Para encurtar, pode ser chamado de método da janela em algumas literaturas. Na análise estatística e de séries temporais pode ser chamado de lag ou método lag. O número de passos prévios é chamado de largura de janela ou tamanho do lag. Essa janela deslizante é a base de como nós podemos transformar qualquer dataset de série temporal em um problema de aprendizado supervisionado. Desse exemplo simples é possível notar algumas coisas:

- Nós podemos ver como isso pode funcionar para transformar uma série temporal em um problema de aprendizado supervisionado, ou de regressão ou um problema de classificação, para valores de séries temporais reais ou classificados (labeled).
- Podemos ver que um dataset de série temporal preparado dessa forma permite a aplicação de algoritmos de machine learning lineares e não-lineares, enquanto a ordem das linhas for preservada.
- Podemos ver como a largura da janela deslizante pode aumentar para incluir mais passos prévios.
- Podemos ver como a abordagem de janela deslizante pode ser usada em séries temporais que tem mais de um valor, isso é, séries temporais multivariada.

Iremos explorar alguns desses usos da janela deslizante, começando a seguir usando ela para lidar com séries temporais com mais de uma observação em cada passo de tempo, chamada de série temporal multivariada.

#### Janela Deslizante com Multivariadas

O número de observações gravadas para um determinado tempo em um dataset de série temporal importa. Tradicionalmente, diferentes nomes são empregados:

- Séries Temporais Univariadas: Datasets onde há uma observação em cada passo de tempo, por exemplo temperatura por hora. O exemplo anterior era univariado.
- Séries Temporais Multivariadas: Datasets onde duas ou mais variáveis são observadas em cada passo.

A maioria dos métodos de análise e até livros sobre séries temporais focam em séries univariadas, por serem mais fáceis de utilizar e simples de entender e trabalhar. Multivariadas são mais difíceis de se trabalhar e modelar, além de ter um desempenho ruim com os métodos clássicos.

Análise de séries temporais multivariadas considera simultâneamente múltiplas séries temporais, é, em geral, muito mais complicada do que a análise de séries temporais univariadas.

O ponto certo de se usar machine learning para séries temporais é quando métodos clássicos decaem. Isso se dá com séries temporais univariadas complexas, e é mais provável com multivariadas devido sua complexidade adicional. Abaixo está um exemplo de série temporal multivariada com janela deslizante. Temos o dataset com duas observações em cada passo de tempo. Só estamos interessados em prevêr a **medida2**.

tempo, medida1, medida2

1, 0.2, 88

2, 0.5, 89

3, 0.7, 87

4, 0.4, 88

5, 1.0, 90

Podemos rearranjar esse dataset como um problema de aprendizado supervisionado com uma largura de janela de um. Isso significa que nós vamos usar os valores do passo prévio de **medida1** e **medida2**. Nós também temos disponível o próximo passo de **medida1**. Iremos então prever o próximo passo de **medida2**. Isso nos dá 3 variáveis de entrada e uma de saída.

X1, X2, X3, y

?, ?, 0.2, 88

0.2, 88, 0.5, 89

0.5, 89, 0.7, 87

0.7, 87, 0.4, 88

0.4, 88, 1.0, 90

1.0, 90, ?, ?

A maioria dos modelos de supervised learning precisam ser modificados para poder prever múltiplos valores de saída, porém **redes neurais** conseguem facilmente. Prever mais de um valor é denominado **prever uma sequência**. No caso acima estávamos prevendo duas variáveis de saída diferentes, mas é possível também prever múltiplos passos de uma mesma variável de saída. Isso é chamado **previsão de múltiplos passos**, e será abordado na próxima seção.

#### Janela Deslizante com Múltiplos Passos

O número de passos a serem previstos é importante. Novamente, é tradicional usar nomes diferentes para o problema dependendo do número de passos a serem previstos:

- One-step Forecast: Quando o próximo passo (**t+1**) é previsto.
- Multi-step Forecast: Quando dois ou mais passos futuros deverão ser previstos.

Todos exemplos anteriores foram one-step forecasts. Tem diversas formas de modelar multi-step como um problema de supervised learning. Agora vamos focar com janela deslizante. Considere o mesmo dataset do primeiro exemplo:

time, measure

1, 100

2, 110

3, 108

4, 115

5, 120

Podemos modelar esses dados para prever dois passos, com largura de janela de um:

X1, y1, y2

? 100, 110

100, 110, 108

110, 108, 115

108, 115, 120

115, 120, ?

120, ?, ?

De novo as linhas com ? não podem ser usadas. Nota-se o fardo nas variáveis de entrada, somente com x1 deve-se prever y1 e y2. Experimentação e reflexão é necessário para definir uma largura de janela que resulte em uma performance aceitável.

### Capítulo 5: Basic Feature Engineering

Dados de séries temporais devem ser reenquadrados como um dataset de supervised learning antes que se possa utilizar algoritmos de machine learning. Não existe o conceito de entradas e saídas em séries temporais. Ao invés disso, nós precisamos escolher a variável que desejamos prever e usar **feature engineering** para construir todas as entradas que serão utilizadas para fazer a previsão dessa variável.

É preciso transformar um dataset de série temporal em um problema de supervised learning. Segue exemplo:

Série Temporal

time 1, value 1

time 2, value 2


time 3, value 3

Problema de Supervised Learning

input 1, output 1

input 2, output 2

input 3, output 3

Em machine learning, variáveis de entrada também são chamadas de **features**. A nossa tarefa é criar ou inventar novas features de entrada a partir do dataset de time series. Idealmente só queremos utilizar input features que melhor auxiliam na modelagem da relação entre as entradas (X) e as saídas (y) do que nós queremos prever. Nesse tutorial vamos ver três classes de features que podemos criar a partir do nosso time series dataset:

- **Date Time Features**: são componentes do próprio passo de cada observação.
- **Lag Features**: são valores de um tempo anterior.
- **Window Features**: sumário dos valores de uma janela fixa de passos prévios.

Antes de adentrar nos métodos de criação de input features, vamos revisar o objetivo de feature engineering

#### O Objetivo de Feature Engineering

O objetivo de feature engineering é fornecer uma relação simples e forte entre os novos input features e os output features para o algoritmo de supervised learning poder modelar. Como não há o variáveis de entrada e saída nas séries temporais, é preciso criar elas do zero. Vamos focar em séries temporais univariadas nos exemplos, mas esses métodos são aplicáveis nos problemas multivariados também. A seguir, vamos ver o dataset que iremos utilizar nesse tutorial.

#### Minimum Daily Temperatures Dataset

Esse dataset descreve a temperatura diária mínima da cidade de Melbourne, Australia, durante 10 anos. 

#### Date Time Features

Vamos começar com a feature mais simples que podemos utilizar. Esses são features a partir do date/time de cada observação. Duas features que podemos começar são o integer do mês e do dia de cada observação. Nós podemos imaginar que os algoritmos de supervised learning talvez possam usar essas entradas para ajudar a provocar informações de sazonalidade do tipo time-of-year ou time-of-month. O problema de supervised learning que estamos propondo é prever a temperatura mínima diária de um determinado mês e dia, como segue:

Month, Day, Temperature

Month, Day, Temperature

Month, Day, Temperature

Podemos fazer isso com Pandas. Primeiro a time series é carregada no Pandas como uma Series. Depois criamos um DataFrame novo no Pandas para o dataset que será transformado. Depois disso, cada coluna é adicionada uma de cada vez, onde a informação de mês e dia é extraída da informação time-stamp para cada observação. Abaixo o código que faz isso:

```python
# create date time features of a dataset
from pandas import read_csv
from pandas import DataFrame
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,
parse_dates=True, squeeze=True)
dataframe = DataFrame()
dataframe['month'] = [series.index[i].month for i in range(len(series))]
dataframe['day'] = [series.index[i].day for i in range(len(series))]
dataframe['temperature'] = [series[i] for i in range(len(series))]
print(dataframe.head(5))
```

Usando só essas informações sozinhas podem resultar em um modelo pobre, porém agrupando elas com mais alguns engineered features, pode resultar em um modelo melhor. Desses exemplos é possível ver que não está restrito a somente valores integer. Pode-se utilizar flags binárias também, como por exemplo se a observação foi feita em um feriado público. No caso de um dataset de temperaturas mínimas, talvez a temporada seja mais relevante. Ao criar features específicas como essa que adiciona-se valor ao modelo. Features baseados em date-time são um bom início, mas normalmente serão mais úteis se incluir esses valores como previous time steps. Esses são chamados de lagged values e iremos ver como adicionar essa feature na próxima seção.

#### Lag Features 

Lag Features são o jeito clássico do qual problemas de previsão de séries temporais são transformados em problemas de supervised learning. A abordagem mais simples é prever o valor do próximo tempo (**t+1**) dado um valor no tempo atual (**t**). O problema de supervised learning com valores trocados fica como segue:

Value(t), Value(t+1)

Value(t), Value(t+1)

Value(t), Value(t+1)

A biblioteca Pandas fornece a função **shift()** para ajudar a criar essas lag features trocadas a partir de um dataset de time series. Pulando o dataset em 1 cria-se a coluna **t**, adicionando um valor **NaN**(desconhecido) na primeira linha. O dataset sem essa troca representa **t+1**. Vamos ver um exemplo. Os primeiros 3 valores do dataset de temperatura são 20.7, 17.9, e 18.8. A lista de valores trocados e não trocados para as 3 primeiras observações são portanto:

Shifted, Original

NaN, 20.7

20.7, 17.9

17.9, 18.8

Podemos concatenar as colunas trocadas para juntá-las em um novo **DataFrame** usando a função **concat()** junto da coluna de eixo (**axis=1**). Juntando tudo isso, abaixo segue um exemplo de criação de uma lag feature para nosso dataset de temperaturas diárias. Os valores são extraídos da series carregada e a lista de valores trocados e não trocados é criada. Cada coluna também é nomeada no DataFrame para prover clareza.

```python
# criando uma lag feature
from pandas import read_csv
from pandas import DataFrame
from pandas import concat
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,
                 parse_dates=True, squeeze=True)
temps = DataFrame(series.values)
dataframe = concat([temps.shift(1), temps], axis=1)
dataframe.columns = ['t', 't+1']
print("Time Series")
print(series.head(5))
print("Valores da Time Series")
print(temps.head(5))
print("Lagged Feature")
print(dataframe.head(5))
```

Pode-se ver que teremos que descartar a primeira linha já que ela não tem informação o suficiente. A adição de lag features é chamada de sliding window method, nesse caso com uma largura de janela de 1. É como se estivessemos deslizando nosso foco sobre a time series para cada observação, com um interesse somente no que está dentro da largura da janela. Podemos expandir essa largura de janela para incluir mais lag features. Por exemplo, abaixo é o caso acima modificado para incluir os últimos 3 valores observados para prever o valor do próximo passo.

```python
# criando lag features
from pandas import read_csv
from pandas import DataFrame
from pandas import concat
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,
                 parse_dates=True, squeeze=True)
temps = DataFrame(series.values)
dataframe = concat([temps.shift(3), temps.shift(2), temps.shift(1), temps], axis=1)
dataframe.columns = ['t-2', 't-1', 't', 't+1']
print(dataframe.head(5))
```

Novamente percebe-se que é preciso descartar as primeiras linhas, pois não tem dados suficientes para treinar o modelo. A dificuldade com a abordagem de janela deslizante é quão grande deve ser a janela para o seu problema. Um bom ponto de partida é fazer uma sensitivity analysis e tentar um conjunto de diferentes larguras de janela pra ver qual se adequa melhor ao problema.

Mas, por que parar com uma janela linear? Talvez você precise de um valor de lag da semana, mês ou ano passado. De novo, isso cai em um domínio específico. No caso do dataset de temperatura, um valor de lag de um mesmo dia do ano ou mês passado pode ser útil. Podemos fazer mais com uma janela do que incluir valores crus. Na próxima seção, vamos ver como incluir features que sumarizam estatísticas através da janela.

#### Rolling Window Statistics

Um passo além de adicionar valores de lag crus é adicionar um sumário dos valores de um tempo prévio. Podemos calcular essa estatística de sumário sobre os valores dentro da janela deslizante e incluir essas features no nosso dataset. Talvez o mais útil seja a média dos valores prévios, também chamado de média rolante (rolling mean).

Podemos calcular a média do valor atual e do prévio e usar isso para prever o próximo valor. Para os dados de temperatura, nós deveriamos esperar 3 passos de tempo antes de ter 2 valores para pegar a média, para usar esses valores para prever o terceiro valor. Por exemplo:

mean(t-1, t), t+1

mean(20.7, 17.9), 18.8

19.3, 18.8

Pandas possui uma função chamada **rolling()** que cria uma nova estrutura de dados com os valores da janela de cada time step. Podemos realizar funções de estatística na janela de valores coletada para cada time step, como calcular a média. Primeiro, a série deve ser *shiftada*. Depois o rolling dataset pode ser criado e os valores de média calculados para cada janela de dois valores. Aqui estão os valores nas primeiras três rolling windows:

#, Window Values

1, NaN

2, NaN, 20.7

3, 20.7, 17.9

Isso sugere que não teremos dados usáveis até a linha 3. Finalmente, como na seção anterior, podemos usar o **concat()** para construir um novo dataset somente com nossas novas colunas. O exemplo abaixo demonstra como fazer isso com Pandas com uma janela de tamanho 2.

```python
# create a rolling mean feature
from pandas import read_csv
from pandas import DataFrame
from pandas import concat
from pandas import to_numeric
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,
parse_dates=True, squeeze=True)
temps = DataFrame(series.values)
shifted = temps.shift(1)
window = shifted.rolling(window=2)
means = window.mean()
dataframe = concat([means, temps], axis=1)
dataframe.columns = ['mean(t-1,t)', 't+1']
print(dataframe.head(5))
```

Rodando o exemplo imprime as primeiras 5 linhas do novo dataset. Podemos ver que as primeiras duas linhas não são úteis.

- A primeira NaN foi criada pelo shift da série.
- A segunda NaN foi criada por que não pode ser usada para calcular uma média.
- Finalmente, a terceira linha mostra o valor esperado de 19.30 (a média de 20.7 e 17.9) usado para prever o terceiro valor na série, 18.8

Tem mais estatísticas que podemos calcular, e até formas matemáticas diferentes de se calcular a definição da *janela*. Abaixo está um exemplo que mostra uma largura de janela de 3 e um dataset composto de mais estatísticas, especificamente o valor mínimo, a média e o valor máximo na janela.

Pode-se ver no código que estamos explicitamente especificando a largura da janela deslizante como uma variável nomeada. Isso nos possibilita utilizá-la tanto para calcular o shift correto como para especificar a largura da janela para a função **rolling()**.

Nesse caso, a largura de janela de 3 significa que devemos shiftar a series para frente em 2 passos de tempo. Isso faz com que as duas primeiras fileiras sejam NaN. Depois, precisamos calcular as estatísticas da janela com 3 valores por janela. É necessário 3 linhas antes de termos dados suficiente da series na janela para começar a calcular as estatísticas. Os valores das 5 primeiras janelas são como segue:

#, Window Values

1, NaN

2, NaN, NaN

3, NaN, NaN, 20.7

4, NaN, 20.7, 17.9

5, 20.7, 17.9, 18.8

Isso sugere que não teremos dados úteis até pelo menos a linha 5 (index 4 do array)

```python
# create rolling statistics features
from pandas import read_csv
from pandas import DataFrame
from pandas import concat
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,
                 parse_dates=True, squeeze=True)
temps = DataFrame(series.values)
width = 3
shifted = temps.shift(width - 1)
window = shifted.rolling(window=width)
dataframe = concat([window.min(), window.mean(), window.max(), temps], axis=1)
dataframe.columns = ['min', 'mean', 'max', 't+1']
print(dataframe.head(5))
```

#### Expandindo Estatísticas de Janela

Um outro tipo de janela é o *expanding window*, que inclui todos os dados prévios da série. Pandas possui uma função chamada **expanding()**, que coleta conjuntos de todos os valores de passos prévios. Essa lista de numeros prévios pode ser sumarizada e incluída como uma nova feature. Abaixo segue a lista de números da expanding window para os primeiros 5 passos de tempo da série:

#, Window Values

1, 20.7

2, 20.7, 17.9,

3, 20.7, 17.9, 18.8

4, 20.7, 17.9, 18.8, 14.6

5, 20.7, 17.9, 18.8, 14.6, 15.8

É necessário dar shift de 1 passo para que o valor que se deseja prever não seja incluído. Por fim, a janela de entradas fica como segue:

#, Window Values

1, NaN

2, NaN, 20.7

3, NaN, 20.7, 17.9,

4, NaN, 20.7, 17.9, 18.8

5, NaN, 20.7, 17.9, 18.8, 14.6

Felizmente os calculos de estatística excluem o NaN e não precisa fazer mais nada. Abaixo segue exemplo do cálculo do valor mínimo, máximo e da média, numa expanding window do dataset de temperaturas diárias.

```python
# create expanding window features
from pandas import read_csv
from pandas import DataFrame
from pandas import concat
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,
                 parse_dates=True, squeeze=True)
temps = DataFrame(series.values)
window = temps.expanding()
dataframe = concat([window.min(), window.mean(), window.max(), temps.shift(-1),], axis=1)
dataframe.columns = ['min', 'mean', 'max', 't+1']
print(dataframe.head(5))
```

#### Lag Scatter Plots

Um tipo útil de gráfico para explorar a relação entre cada observação e um lag associado a ela é chamado de *scatter plot*. Pandas tem uma função para isso chamada de lag plot. Ele coloca a observação do tempo **t** no eixo X e a observação do próximo tempo (**t+1**) no eixo Y.

- Se os pontos se agrupam sobre uma linha diagonal do canto esquerdo inferior para o canto direito superior, isso sugere uma correlação positiva.
- Se os pontos se agrupam sobre uma linha diagonal do canto esquerdo superior para o canto direito inferior, isso sugere uma correlação negativa.
- Ambas são boas pois podem ser modeladas.

Mais pontos apertados na linha diagonal sugerem uma relação forte e mais separados da linha sugerem uma relação fraca. Uma bola no meio ou um espalhamento sobre o gráfico sugerem uma relação fraca ou inexistente. Abaixo um exemplo de lag plot.

```python
# create a scatter plot
from pandas import read_csv
from matplotlib import pyplot
from pandas.plotting import lag_plot
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,
                 parse_dates=True, squeeze=True)
lag_plot(series)
pyplot.show()
```

Esse gráfico mostra uma relação positiva e forte entre as observações e seus valores **lag1**.

Pode-se repetir o processo para qualquer valor de lag. Talvez com a observação de um mesmo tempo na semana passada, ou mês passado, ou ano passado, ou qualquer outro conhecimento de domínio específico. Por exemplo, podemos criar um scatter plot para a observação de cada valor nos últimos 7 dias. Primeiro um novo **DataFrame** é criado com valores de lag como novas colunas. As colunas são nomeadas de acordo. Então um novo subplot é criado onde "ploteia" cada observação com um valor de lag diferente.

```python
# create multiple scatter plots
from pandas import read_csv
from pandas import DataFrame
from pandas import concat
from matplotlib import pyplot
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,
parse_dates=True, squeeze=True)
values = DataFrame(series.values)
lags = 7
columns = [values]
for i in range(1,(lags + 1)):
    columns.append(values.shift(i))
dataframe = concat(columns, axis=1)
columns = ['t']
for i in range(1,(lags + 1)):
    columns.append('t-' + str(i))
dataframe.columns = columns
pyplot.figure(1)
for i in range(1,(lags + 1)):
    ax = pyplot.subplot(240 + i)
    ax.set_title('t vs t-' + str(i))
    pyplot.scatter(x=dataframe['t'].values, y=dataframe['t-'+str(i)].values)
pyplot.show()
```

#### Autocorrelation Plots

Podemos quantificar a força e o tipo de relação entre observações e seus lags. Em estatística é chamado de correlação, e quando calculado sobre valores de lag em séries temporais, é chamado de autocorrelação. A correlação entre dois grupos de números como observações e seus valores de **lag=1**, resultam em um número entre -1 e 1. O sinal desse numero indica uma relação negativa ou positiva respectivamente. Um valor próximo de zero sugere uma correlação fraca, enquanto um valor próximo de -1 ou 1 sugere uma correlação forte.

Valores de correlação, chamados de coeficientes de correlação, podem ser calculados para cada observação e valores de lag diferentes. Assim que são calculados podem ser plotados pelo Pandas a partir da função **autocorrelation_plot()**. Segue exemplo disso:

```python
# create an correlation plot
from pandas import read_csv
from matplotlib import pyplot
from pandas.plotting import autocorrelation_plot
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,
                 parse_dates=True, squeeze=True)
autocorrelation_plot(series)
pyplot.show()
```

Os valores de lag estão no eixo X e a correlação no eixo Y. Podemos ver no dataset de temperaturas ciclos de uma forte correlação negativa e positiva. Isso captura a relação de uma observação com observações passadas de uma mesma ou oposta estação do ano. Ondas como essas são um forte sinal de sazonalidade.

### Capítulo 7: Reamostragem e Interpolação

As vezes as observações estão numa frequência errada, são muito ou pouco granulares. Pandas pode mudar a frequência dos dados de time series, aumentando ou diminuindo.

#### Resampling

Resampling envolve mudar a frequência das observações de time series. Dois tipos de resampling são:

- **Upsampling**: Onde você aumenta a frequência das amostras, como por exemplo de minutos para segundos.
- **Downsampling**: Onde você diminui a frequência das amostras, como por exemplo de dias para meses.

Para upsampling precisa determinar como as observações serão calculadas com interpolação, no downsampling precisa selecionar as estatísticas sumarizadas para calcular os novos valores agregados.

Dois motivos principais para se interessar pelo resampling:

- **Problem Framing**: Quando os dados não estão disponíveis na mesma frequência da qual você quer fazer previsões.
- **Feature Engineering**: Para providenciar estrutura adicional ou insight no learning problem para supervised learning models.

#### Shampoo Sales Dataset

Nessa aula usaremos o Shampoo Sales Dataset, descreve o número de vendas de shampoo em 3 anos. Bote no diretório com o nome **shampoo-sales.csv**. Como o dataset só tem um número de ano e meses, foi necessário uma função customizada para parsear as datas e definir o ano como um arbitrário (1900) para que o date-times pudesse ser interpretado pelo Pandas **Series**.

#### Upsampling Data

As observações desse dataset estão por mês, caso seja necessário informação diária, será preciso realizar o upsampling de mês para dia, e usar um esquema de interpolação para preencher essa nova frequência diária. Pandas possui uma função chamada **resample()** nos objetos **Series** e **DataFrame**. Pode ser usado para agrupar os registros quando fazendo downsampling e cria espaço para novas observações quando fazendo upsampling.

```python
# upsample to daily intervals
from pandas import read_csv
from pandas import datetime
def parser(x):
    return datetime.strptime('190'+x, '%Y-%m')
series = read_csv('shampoo-sales.csv', header=0, index_col=0, parse_dates=True,
    squeeze=True, date_parser=parser)
upsampled = series.resample('D').mean()
print(upsampled.head(32))
```

Podemos ver que a função criou novos valores NaN para cada dia novo que ela criou, podemos utilizar a função **interpolate()** para interpolar esses valores que faltam. Diversos tipos de interpolação estão disponíveis, para simplificar podemos começar com uma interpolação linear. Desenha uma linha reta entre os dados disponíveis, nesse caso no primeiro do mês, e preenche os valores na frequência desejada a partir dessa linha.

```python
# upsample to daily intervals with linear interpolation
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot

def parser(x):
    return datetime.strptime('190'+x, '%Y-%m')

series = read_csv('shampoo-sales.csv', header=0, index_col=0, parse_dates=True,
                 squeeze=True, date_parser=parser)
upsampled = series.resample('D').mean()
interpolated = upsampled.interpolate(method='linear')
print(interpolated.head(32))
interpolated.plot()
pyplot.show()
```

É o mesmo gráfico do dataset original, não muda pois a interpolação só preencheu com valores para tracejar a linha.

Outra interpolação comum é utilizar uma polinomial ou *spline* para conectar os valores, isso cria mais curvas e pode parecer mais natural em muitos datasets. Para isso é necessário especificar a ordem (número de termos no polinomial). Nesse caso ordem de 2 é suficiente.

```python
# upsample to daily intervals with spline interpolation
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot

def parser(x):
    return datetime.strptime('190'+x, '%Y-%m')

series = read_csv('shampoo-sales.csv', header=0, index_col=0, parse_dates=True,
                 squeeze=True, date_parser=parser)
upsampled = series.resample('D').mean()
interpolated = upsampled.interpolate(method='spline', order=2)
print(interpolated.head(32))
interpolated.plot()
pyplot.show()
```

Geralmente interpolação é útil quando está faltando observações.

#### Downsampling Data

Os dados estão por mês, mas talvez seja interessante dividir por trimestre. Invés de criar novas linhas entre observações existentes, a função **resample()** do Pandas irá agrupar todas observações de acordo com a nova frequência. Pandas tem um alias para trimestre, **Q**. Precisa pensar como criar um novo valor trimestral para cada group de 3 registros. Para isso pode-se usar a função **mean()**.

```python
# downsample to quarterly intervals
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot
def parser(x):
    return datetime.strptime('190'+x, '%Y-%m')

series = read_csv('shampoo-sales.csv', header=0, index_col=0, parse_dates=True,
                 squeeze=True, date_parser=parser)
resample = series.resample('Q')
quarterly_mean_sales = resample.mean()
print(quarterly_mean_sales.head())
quarterly_mean_sales.plot()
pyplot.show()
```

É possível também transformar de mês para ano usando o alias **A**, depois disso aplicar uma soma para calcular o total de vendas cada ano.

```python
#downsample to yearly intervals
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot

def parser(x):
    return datetime.strptime('190'+x, '%Y-%m')

series = read_csv('shampoo-sales.csv', header=0, index_col=0, parse_dates=True,
                 squeeze=True, date_parser=parser)
resample = series.resample('A')
yearly_mean_sales = resample.sum()
print(yearly_mean_sales.head())
yearly_mean_sales.plot()
pyplot.show()
```

### Capítulo 8: Power Transforms

Data transforms procuram reduzir ruído e melhorar o sinal na previsão de time series.

#### Airline Passengers Dataset

Nessa aula usaremos o dataset Airline Passengers, ele descreve o número total de passageiros de uma airline em um período de tempo. coloque no diretório com o nome **airline-passengers**. Exemplo abaixo carrega o dataset e plota os dados.

#### Square Root Transform

Uma série temporal que tem tendência de crescimento quadrático pode se tornar linear ao tirar a raíz quadrada. Por exemplo, considere números de 1 a 99 ao quadrado, o gráfico linear dessa série irá mostrar uma tendência de crescimento quadrático e um histograma irá mostrar uma distribuição exponencial com uma longa cauda. Segue exemplo:

```python
# contrive a quadratic time series
from matplotlib import pyplot
series = [i**2 for i in range(1,100)]
pyplot.figure(1)
#line plot
pyplot.subplot(211)
pyplot.plot(series)
# histogram
pyplot.subplot(212)
pyplot.hist(series)
pyplot.show()
```

Aplicando a raíz quadrada a partir do **sqrt()**, o gráfico fica linear e o histograma fica com distribuição uniforme.

```python
# square root transform a contrived quadratic time series
from matplotlib import pyplot
from numpy import sqrt
series = [i**2 for i in range(1,100)]
# sqrt transform
transform = series = sqrt(series)
pyplot.figure(1)
# line plot
pyplot.subplot(211)
pyplot.plot(transform)
# histogram
pyplot.subplot(212)
pyplot.hist(transform)
pyplot.show()
```

Esse comportamento é esperado de acontecer no dataset airline passengers, se ele tiver mesmo um crescimento quadrático. uma **square root transform** irá reduzir a tendência de crescimento para ficar linear e mudar a distribuição para talvez próxima de uma normal. O exemplo abaixo aplica esse transform no dataset e plota os resultados.

```python
# square root transform in a time series
from pandas import read_csv
from pandas import DataFrame
from numpy import sqrt
from matplotlib import pyplot
series = read_csv('airline-passengers.csv', header=0, index_col=0, parse_dates=True,
                 squeeze=True)
dataframe = DataFrame(series.values)
dataframe.columns = ['passengers']
dataframe['passengers'] = sqrt(dataframe['passengers'])
pyplot.figure(1)
# line plot
pyplot.subplot(211)
pyplot.plot(dataframe['passengers'])
# histogram
pyplot.subplot(212)
pyplot.hist(dataframe['passengers'])
pyplot.show()
```

É possível ver que a tendência foi reduzida porém não extinta, pois ainda mostra uma variação crescente de ciclo em ciclo e uma longa cauda na direita do histograma, que sugere uma distribuição exponencial.

#### Log Transform

Abaixo um exemplo do log transform pra fazer a distribuição ficar uniforme e a série linear.

```python
# log transform a contrived exponential time series
from matplotlib import pyplot
from math import exp
from numpy import log
series = [exp(i) for i in range(1,100)]
transform = log(series)
pyplot.figure(1)
# line plot
pyplot.subplot(211)
pyplot.plot(transform)
# histogram
pyplot.subplot(212)
pyplot.hist(transform)
pyplot.show()
```

#### Box-Cox Transform

Box-Cox é um método que avalia qual o melhor power transform para uma determinada série, ele é configurável e possuí diversas transforms. Uma implementação está presente na biblioteca **scipy.stats**. A função **boxcox()** pega um argumento, chamado **lambda**, que controla o tipo de transform a ser performada. Abaixo alguns valores comuns para **lamba**:

- lambda = -1.0 is a reciprocal transform.
- lambda = -0.5 is a reciprocal square root transform.
- lambda = 0.0 is a log transform.
- lambda = 0.5 is a square root transform.
- lambda = 1.0 is no transform.

Exemplo de log transform com **boxcox**:

```python
# manually box-cox transform a time series
from pandas import read_csv
from pandas import DataFrame
from scipy.stats import boxcox
from matplotlib import pyplot
series = read_csv('airline-passengers.csv', header=0, index_col=0, parse_dates=True,
                 squeeze=True)
dataframe = DataFrame(series.values)
dataframe.columns = ['passengers']
dataframe['passengers'] = boxcox(dataframe['passengers'], lmbda=0.0)
pyplot.figure(1)
# line plot
pyplot.subplot(211)
pyplot.plot(dataframe['passengers'])
# histogram
pyplot.subplot(212)
pyplot.hist(dataframe['passengers'])
pyplot.show()
```

Podemos deixar o valor de **lambda** sem nada (padrão) e deixar a função achar um valor otimizado. O exemplo abaixo demonstra esse uso, retornando tanto o dataset transformado e o valor **lambda**.

```python
# automatically box-cox transform a time series
from pandas import read_csv
from pandas import DataFrame
from scipy.stats import boxcox
from matplotlib import pyplot
series = read_csv('airline-passengers.csv', header=0, index_col=0, parse_dates=True,
                 squeeze=True)
dataframe = DataFrame(series.values)
dataframe.columns = ['passengers']
dataframe['passengers'], lam = boxcox(dataframe['passengers'])
print('Lambda: %f' % lam)
pyplot.figure(1)
# line plot
pyplot.subplot(211)
pyplot.plot(dataframe['passengers'])
# histogram
pyplot.subplot(212)
pyplot.hist(dataframe['passengers'])
pyplot.show()
```

Lambda de 0.14 está mais próximo de 0.0 que é log e longe de 0.5 que é o square root.

### Capítulo 9: Moving Average Smoothing

Moving average smoothing é uma técnica efetiva que pode ser usada para feature engineering, preparação dos dados e até previsões. A seguir vamos ver o que é isso e como usar para esses 3 casos.

#### Moving Average Smoothing

Smoothing é uma técnica aplicada a time series para remover as variações entre os time steps. A esperança é que seja removida o ruído. Moving Average é um tipo comum e simples de smoothing, e involve criar uma nova series onde os valores abrangem a média dos valores originais. Ele utiliza uma janela deslizante para definir o número de observações que será usada para calcular a média móvel. Existem dois tipos: Centered e Trailing Moving Average.

#### Centered Moving Average

O valor no tempo (**t**) é calculado como a média do valor passado, presente e futuro, um exemplo com janela de largura 3 seria:

center ma(t) = mean(obs(t - 1); obs(t); obs(t + 1))

Para esse método é necessário saber os valores futuros, por isso ele é usado em análise, e não em previsão.

#### Trailing Moving Average

O valor no tempo (**t**) é calculado como a média do valor passado e presente. Um exemplo com janela de largura 3 seria:

trail ma(t) = mean(obs(t - 2); obs(t - 1); obs(t))

Utiliza somente valores históricos e pode ser usado para previsão, será o foco desse tutorial.

#### Data Expectations

Para calcular a média móvel de uma time series é preciso fazer algumas suposições sobre os dados, assumindo que eles não tenham tendências nem sazonalidades. Existe diversos métodos para remover esses dois componentes, dois deles para cada seria usar o differencing method e modelar o comportamento para subtrair explicitamente da série.

#### Moving Average as Data Preparation

Moving Average pode ser usada para preparar os dados ao reduzir a variação randômica deles e assim melhor expôr sua estrutura.

Pandas possui a função **rolling()** que automaticamente agrupa as observações em uma janela. Pode especificar o tamanho da mesma, e por padrão uma trailing window é criada. Assim que a janela é criada é possivel pegar o valor da média para ser o dataset transformado.

obs(t) = 1/3 X (obs(t - 2) + obs(t - 1) + obs(t))

Para o **dataset de Daily Female Births**, a primeira média móvel seria em 3 de janeiro, como segue:

obs(t) = 1/3 X (obs(t - 2) + obs(t - 1) + obs(t)) = 1/3 X (35 + 32 + 30) = 32.333

Abaixo um exemplo de transformar o dataset em uma média móvel com janela de tamanho de 3 dias, escolhidos aleatóriamente:

```python
# moving average smoothing as data preparation
from pandas import read_csv
from matplotlib import pyplot
series = read_csv('daily-total-female-births.csv', header=0, index_col=0, parse_dates=True,
                 squeeze=True)
# tail-rolling average transform
rolling = series.rolling(window=3)
rolling_mean = rolling.mean()
print(rolling_mean.head(10))
# plot original and transformed dataset
series.plot()
rolling_mean.plot(color='red')
pyplot.show()
# zoomed plot original and transformed dataset
series[:100].plot()
rolling_mean[:100].plot(color='red')
pyplot.show()
```

#### Moving Average as Feature Engineering

Para problemas de supervised learning, é possível adicionar a média móvel como uma nova input feature, usada para prever o próximo time step. Primeiro, uma cópia da série precisa ser shiftada para frente por um time step. Isso irá representar a entrada para o problema de previsão, ou uma versão **lag=1** da série. Segue exemplo:

X, y

NaN, obs1

obs1, obs2

obs2, obs3

Depois, uma segunda cópia precisa ser shiftada para frente por um, menos o tamanho da janela. Para que a média móvel sumarize os últimos valores e não inclua o valor a ser previsto.

Por exemplo, numa janela de tamanho **3**, precisa shiftar a série **2** passos pra frente. Por que precisa incluir as duas observações prévias e a observação atual para prever o próximo valor. Depois disso pode-se calcular a média móvel dessa série shiftada. Abaixo um exemplo de como as primeiras 5 linhas de valores de média móvel ficariam.

Observations, Mean

NaN NaN

NaN, NaN NaN

NaN, NaN, 35 NaN

NaN, 35, 32 NaN

30, 32, 35 32

Abaixo um exemplo de inclusão da média móvel dos 3 valores prévios como uma nova feature, além da input feature original **t**.

```python
# moving average smoothing as feature engineering
from pandas import read_csv
from pandas import DataFrame
from pandas import concat
series = read_csv('daily-total-female-births.csv', header=0, index_col=0, parse_dates=True,
                 squeeze=True)
df = DataFrame(series.values)
width = 3
lag1 = df.shift(1)
lag3 = df.shift(width - 1)
window = lag3.rolling(window=width)
means = window.mean()
dataframe = concat([means, lag1, df], axis=1)
dataframe.columns = ['mean', 't', 't+1']
print(dataframe.head(10))
```

#### Moving Average as Prediction

O valor da média móvel pode ser usada para previsão, é um naive model que assume que tendências e sazonalidades foram tratadas. Esse modelo pode ser usado em um modo *walk-forward*. Assim que novas observações ficam disponíveis (ex diariamente), o modelo pode ser atualizado e uma previsão é feita para o próximo dia. Pode ser implementado manualmente em Python. Abaixo um exemplo de um modelo de média móvel usado em um walk-forward manner.

```python
# moving average smoothing as a forecast model
from math import sqrt
from pandas import read_csv
from numpy import mean
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot
series = read_csv('daily-total-female-births.csv', header=0, index_col=0, parse_dates=True,
                 squeeze=True)
# prepare situation
X = series.values
window = 3
history = [X[i] for i in range(window)]
test = [X[i] for i in range(window, len(X))]
predictions = list()
# walk forward over time steps in test
for t in range(len(test)):
    length = len(history)
    yhat = mean([history[i] for i in range(length-window,length)])
    obs = test[t]
    predictions.append(yhat)
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
rmse = sqrt(mean_squared_error(test, predictions))
print('Test RMSE: %.3f' % rmse)
# plot
pyplot.plot(test)
pyplot.plot(predictions, color='red')
pyplot.show()
# zoom plot
pyplot.plot(test[:100])
pyplot.plot(predictions[:100], color='red')
pyplot.show()
```

### Capítulo 10: Uma Introdução Gentil ao Ruído Branco

Se a série temporal é ruído branco quer dizer que ela é só números aleatórios e não pode ser prevista. Se os erros de uma previsão não são ruído branco, quer dizer que o modelo pode ser melhorado.

#### O que é um Ruído Branco?

Se os valores da série temporal são independentes e sem relação entre eles, e são identicamente distribuídas, quer dizer que são ruído branco.

#### Por que isso importa?

- **Previsibilidade**: Se a série temporal é ruído branco ela é aleatória e não pode ser modelada e prevista.
- **Diagnóstico**: A série de erros de um modelo de previsão de série temporal idealmente deveria ser ruído branco.

Se os erros de um modelo de previsão são ruído branco, quer dizer que ele está otimizado e esse erro é inerente e não pode ser resolvido. Se não for ruído branco, quer dizer que o modelo ainda pode ser melhorado.

#### Sua série temporal é ruído branco?

A série temporal não é ruído branco se qualquer um das condições abaixo forem verdadeiras:

- A série tem uma média diferente de zero?
- A variância muda no decorrer do tempo?
- Os valores correlacionam com os valores de lag?

Algumas ferramentas para checar se a série é ruído branco são:

- **Criar um line plot**: Checar por features como uma média que muda, variância, ou relacionamentos óbvios entre variáveis lagadas.
- **Calcular estatísticas de sumário**: Checar a média e variância de toda a série contra a média e variância de blocos importantes de valores (ex: dias, meses ou anos).
- **Criar um autocorrelation plot**: Checar por relação entre as variáveis lagadas.

Depois, podemos calcular e imprimir algumas estatísticas, incluindo a média e desvio padrão da série.

Se a série fosse maior seria ideal dividir ela em duas e comparar a média e desvio padrão das duas metades, deveria ser igual. Agora podemos criar alguns gráficos:

### Capítulo 11: Uma introdução ao Random Walk

Random Walk é um método que ajuda a descobrir se a série temporal é previsível ou não.

#### Random Series

A biblioteca padrão do Python tem um módulo **random** que suporta funções para gerar números randômicos. A função **randrange()** pode ser usada para criar integer random de 0 até um limite superior. Podemos usar essa função para gerar uma lista de 1000 integers randoms entre 0 e 10. Exemplo como segue:

```python
# create and plot a random series
from random import seed
from random import randrange
from matplotlib import pyplot
seed(1)
series = [randrange(10) for i in range(1000)]
pyplot.plot(series)
pyplot.show()
```

Isso não é Random Walk, é só uma sequência de números randômicos também chamada de ruído branco.

#### Random Walk

Uma random walk é diferente de uma lista de números random porque o próximo valor na sequência é uma modificação do valor anterior. O processo para gerar a série força uma dependência de um passo para o próximo. Um modelo simples de random walk é como segue:

- 1. Comece com um número aleatório de -1 ou 1.
- 2. Aleatóriamente selecione um -1 ou 1 e adicione à observação do passo anterior.
- 3. Repita o passo 2 tanto quanto desejar.

Podemos criar esse processo em Python ao criar um laço com esse processo, criando uma lista de 1000 time steps para o random walk. Exemplo completo como segue:

```python
# create and plot a random walk
from random import seed
from random import random
from matplotlib import pyplot
seed(1)
random_walk = list()
random_walk.append(-1 if random() < 0.5 else 1)
for i in range(1, 1000):
    movement = -1 if random() < 0.5 else 1
    value = random_walk[i-1] + movement
    random_walk.append(value)
pyplot.plot(random_walk)
pyplot.show()
```

Bem diferente de um ruído branco, realmente parece uma time series.

#### Random Walk and Autocorrelation

Podemos calcular a correlação entre cada observação e seu valor anterior. Um gráfico disso é chamado de autocorrelation plot ou correlogram. Do jeito que o random walk é construído, espera-se uma forte correlação entre os valores e uma queda linear. Podemos usar a função do Pandas **autocorrelation_plot()** para plotar o correlogram. O exemplo segue abaixo com a mesma seed dos outros para garantir os mesmos números randômicos.

```python
# plot the autocorrelation of a random walk
from random import seed
from random import random
from matplotlib import pyplot
from pandas.plotting import autocorrelation_plot
seed(1)
random_walk = list()
random_walk.append(-1 if random() < 0.5 else 1)
for i in range(1, 1000):
    movement = -1 if random() < 0.5 else 1
    value = random_walk[i-1] + movement
    random_walk.append(value)
autocorrelation_plot(random_walk)
pyplot.show()
```

#### Random Walk e Estacionariedade

Uma time series estacionário é uma onde os valores não são uma função de tempo. Do jeito que foi criado nosso random walk, sabemos que as observações nele são dependentes do tempo, pois cada uma é calculada a partir do passo anterior. Nesse caso o random walk é não-estacionário (todos são). Nem todas séries temporais não-estacionárias são random walks, e elas normalmente não terão uma média e/ou variância consistente no decorrer do tempo, já o random walk sim. Podemos confirmar isso usando um teste de significância, o teste Augmented Dickey-Fuller.

Podemos realizar esse teste usando a função **adfuller()** da biblioteca Statsmodels:

```python
# calculate the stationary of a random walk
from random import seed
from random import random
from statsmodels.tsa.stattools import adfuller
# generate random walk
seed(1)
random_walk = list()
random_walk.append(-1 if random() < 0.5 else 1)
for i in range(1, 1000):
    movement = -1 if random() < 0.5 else 1
    value = random_walk[i-1] + movement
    random_walk.append(value)
# statistical test
result = adfuller(random_walk)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

Podemos ver que o test statistic value era 0.341605. Isso é maior que todos os valores críticos nos níveis de confiança 1%, 5% e 10%. Com isso, podemos dizer que a série temporal aparenta ser não-estacionária.

Podemos tornar o random walk estacionário ao tirar a diferença entre cada observação e a anterior.

```python
# calculate and plot a differenced random walk
from random import seed
from random import random
from matplotlib import pyplot
# create random walk
seed(1)
random_walk = list()
random_walk.append(-1 if random() < 0.5 else 1)
for i in range(1, 1000):
    movement = -1 if random() < 0.5 else 1
    value = random_walk[i-1] + movement
    random_walk.append(value)
# take difference
diff = list()
for i in range(1, len(random_walk)):
    value = random_walk[i] - random_walk[i-1]
    diff.append(value)
# line plot
pyplot.plot(diff)
pyplot.show()
# autocorrelation plot
autocorrelation_plot(diff)
pyplot.show()
```

Assim observamos uma série de ruído branco sem nenhuma estrutura e impossibilitada de modelagem.

#### Prevendo uma Random Walk

Random walk são imprevisíveis, não podem ser previstas. Dado o jeito como o o random walk é construído, podemos assumir que o melhor jeito de prevê-lo é usando a previsão anterior para definir a próxima, isso devido ao fato de sabermos que os passos são uma função do passo anterior. Isso se chama *naive forecast*, ou modelo de persistência.

Podemos implementar isso em Python ao primeiro dividir o dataset em conjuntos de teste e treinamento, depois usando o modelo de persistência para prever a saída usando um método de previsão rolling. Depois que todas as previsões são coletadas para o conjunto de teste, o root mean squared error (RMSE) é calculado.

```python
#persistence forecasts for a random walk
from random import seed
from random import random
from sklearn.metrics import mean_squared_error
from math import sqrt
# generate the random walk
seed(1)
random_walk = list()
random_walk.append(-1 if random() < 0.5 else 1)
for i in range(1, 1000):
    movement = -1 if random() < 0.5 else 1
    value = random_walk[i-1] + movement
    random_walk.append(value)
# prepare dataset
train_size = int(len(random_walk) * 0.66)
train, test = random_walk[0:train_size], random_walk[train_size:]
# persistence
predictions = list()
history = train[-1]
for i in range(len(test)):
    yhat = history
    predictions.append(yhat)
    history = test[i]
rmse = sqrt(mean_squared_error(test, predictions))
print('Persistence RMSE: %.3f' %rmse)
```

Modelo de persistência é o melhor para prever random walks

#### A time series é random walk?

Para ver se a series é random walk deve-se checar:

- Se a time series apresenta uma dependência temporal forte que decai linearmente ou em algum padrão parecido.
- Se a time series é não-estacionária e ao fazer ela ficar estacionária ela não mostrar nenhuma estrutura nos dados.
- O modelo de persistência provê a melhor fonte de previsões confiáveis.

Se você usar modelo de persistência e não houver sinais de que pode ser melhorado a previsão, é um forte indício de que a série é uma random walk.

### Capítulo 12: Decompondo Dados de Séries Temporais

Decomposição de séries temporais involve pensar numa series como uma combinação de componentes: nível, tendência, sazonalidade e ruído. Isso ajuda a entender melhor os dados e como modelá-los.

#### Componentes de Time Series

Uma abstração importante para selecionar modelos de previsão é quebrar a série temporal em componentes sistemáticos e assistemáticos.

- **Sistemáticos**: Componentes que possuem consistência ou recorrência e podem ser descritos e modelados.
- **Assistemáticos**: Componentes que não podem ser modelados diretamente.

Uma série temporal consiste em três componentes sistemáticos, incluindo nível, tendência e sazonalidade, e um componente assistemático, chamado ruído. Esses componentes são definidos como segue:

- **Nível**: O valor médio da série.
- **Tendência**: O valor que está subindo ou descendo da série.
- **Sazonalidade**: O ciclo de curto prazo que se repete na série.
- **Ruído**: A variação aleatória na série.

#### Combinando Componentes de Time Series

Uma série é um conjunto desses componentes, obrigatoriamente possuem nível e ruído, opcionalmente possuem tendência e sazonalidade. Ajuda pensar nos componentes como uma combinação aditiva ou multiplicativa.

##### Modelo Aditivo

Um modelo aditivo sugere que os componentes são adicionados juntos como segue:

y(t) = Nível + Tendência + Sazonalidade + Ruído

Um modelo aditivo é linear onde as mudanças no decorrer do tempo são feitas consistentemente pela mesma quantidade. Uma tendência linear é uma linha reta. Uma sazonalidade linear tem a mesma frequência (largura dos ciclos) e amplitude (altura dos ciclos).

##### Modelo Multiplicativo

Um modelo multiplicativo sugere que os componentes são multiplicados juntos como segue:

y(t) = Nível X Tendência X Sazonalidade X Ruído

Um modelo multiplicativo é não linear, como quadrático ou exponêncial. Mudanças aumentam ou diminuem no decorrer do tempo. Uma tendência não linear é uma linha curva. Uma sazonalidade não linear tem uma frequência que aumenta ou diminui no decorrer do tempo.

#### Decomposição como Ferramenta

Essa abstração é útil para a análise de time series, pois ela ajuda a entender a complexidade e como melhor capturar esses componentes em um dado modelo. Esses componentes são coisas que devem ser pensadas na hora de preparar os dados e selecionar ou melhorar um modelo.

#### Decomposição Automática de Time Series

A biblioteca Statsmodels suporta uma implementação de um método de decomposição clássico, numa função chamada **seasonal_decompose()**. Ele requer que seja especificado se o modelo é aditivo ou multiplicativo. Ambos irão gerar um resultado, é bom saber qual dos dois a série parece. A função **seasonal_decompose()** retorna um result object. Esse result object contém arrays para acessar quatro pedaços de dados da decomposição.

### Capítulo 13: Usando e Removendo Tendências

Tendência é um aspecto importante, ela é um aumento ou queda crescente ao longo do tempo.

#### Tendência em Time Series

Identificar e entender tendências podem ajudar a melhorar a performance do modelo; abaixo estão alguns motivos:

- **Modelagem mais rápida**: Talvez a presença ou ausência de uma tendência podem tornar a escolha e avaliação de um modelo mais eficiente
- **Problema mais simples**: Talvez seja possível corrigir ou remover a tendência para simplificar a modelagem e melhorar a performance do modelo.
- **Mais dados**: A tendência pode ser usada como informação adicional ao modelo e assim melhorar sua performance.

#### Tipos de Tendência

Existem diversos tipos, duas classes são as principais:

- **Tendências Determinísticas**: Tendências que consistentemente aumentam ou diminuem.
- **Tendências Estocásticas**: Aumentam ou diminuem incosistentemente (random).

Podemos pensar sobre tendências no sentido do escopo das observações:

- **Tendências Globais**: São aplicadas à toda a série temporal.
- **Tendências Locais**: São aplicadas apenas à partes ou subsequências de uma série temporal.

#### Identificando uma Tendência

É possível criar gráficos para identificar tendências óbvias, mas isso é um processo subjetivo, sugere-se criar line plots dos dados e ver se acha tendências óbvias.

#### Removendo uma Tendência

Time series com tendência é chamada de não-estacionária e pode ser modelada. Assim que é modelada pode ser removida do dataset. Um dataset sem tendência é chamado de estacionário.

#### Usando Tendências de Time Series em Machine Learning

Em uma perspectiva de Machine Learning, uma tendência nos dados representa duas oportunidades:

- **Remover informação**: Remover informação sistemática que destorce a relação entre variáveis de entrada e saída.
- **Adicionar informação**: Adicionar informação sistemática que melhora a relação entre variáveis de entrada e saída.

Removendo tendências pode ajudar a preparar os dados, isso é usado em métodos estatísticos porém não é eficiente em modelos de machine learning. Alternativamente pode-se adicionar tendências como uma nova variável de entrada para o modelo de supervised learning, com o intuito de prevêr a variável de saída. Ambos podem ser úteis nos problemas de time series.

#### Shampoo Sales Dataset

Nessa aula vamos usar o dataset Shampoo Sales, ele descreve o número mensal de vendas de shampoo no decorrer de 3 anos. Salvar no diretório com o nome **shampoo-sales.csv**.

#### Remover Tendência ao Diferenciar

Um dos métodos mais simples para destender uma time series é diferenciando, isso é, removendo do valor atual o valor da observação anterior, como segue:

valor(t) = observacao(t) - observacao(t-1)

Isso tem o efeito de remover a tendência de um time series dataset. Podemos criar um dataset diferenciado implementando isso diretamente. Uma nova lista de observações pode ser criada. Abaixo um exemplo.

```python
# detrend a time series using differencing
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot

def parser(x):
    return datetime.strptime('190'+x, '%Y-%m')

series = read_csv('shampoo-sales.csv', header=0, index_col=0, parse_dates=True,
                 squeeze=True, date_parser=parser)
X = series.values
diff = list()
for i in range(1, len(X)):
    value = X[i] - X[i - 1]
    diff.append(value)
pyplot.plot(diff)
pyplot.show()
pyplot.plot(series)
pyplot.show()
```

Para tendências quadráticas, que a mudança na tendência aumenta ou diminui, é possível aplicar essa diferenciação mais de uma vez, como num segundo nível de diferenciação.

#### Removendo Tendência por Model Fitting

Model Fitting é tendar encaixar modelos nos datasets, podem ser usados para identificar tendências e até removê-las. Por exemplo, um modelo linear poderia ser encaixado (fit) no time index para prever a observação. O dataset seria como segue:

X, y

1, obs1

2, obs2

3, obs3

4, obs4

5, obs5

A previsão desse modelo formará uma linha reta que pode ser tirada como a linha de tendência do dataset. Essa previsão pode também ser subtraída do time series original para providenciar uma versão sem tendência do dataset.

valor(t) = observação(t) - previsão(t)

Pode ser usado Polynomial curve fitting e outros modelos não lineares também. Pode-se implementar em Python ao treinar um modelo do scikit-learn, o **LinearRegression**.

```python
# use a linear model to detrend a time series
from pandas import read_csv
from pandas import datetime
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot
import numpy

def parser(x):
    return datetime.strptime('190'+x, '%Y-%m')

series = read_csv('shampoo-sales.csv', header=0, index_col=0, parse_dates=True,
                 squeeze=True, date_parser=parser)
# fit linear model
X = [i for i in range(0, len(series))]
X = numpy.reshape(X, (len(X), 1))
y = series.values
model = LinearRegression()
model.fit(X, y)
# calculate trend
trend = model.predict(X)
# pyplot trend
pyplot.plot(y)
pyplot.plot(trend)
pyplot.show()
# detrend
detrended = [y[i]-trend[i] for i in range(0, len(series))]
# plot detrended
pyplot.plot(detrended)
pyplot.show()
```

Primeiro traça uma linha sobre a tendência, depois remove ela. Se tiver uma parábola nos resíduos quer dizer que um modelo polynomial se encaixaria (fit) melhor.

### Capítulo 14: Usando e Removendo Sazonalidade

Sazonalidade é um ciclo que se repete no decorrer do tempo, como por exemplo mensal ou anualmente. Pode obscurecer o sinal que se gostaria de prever, e pode providenciar um sinal forte para modelos preditivos.

#### Sazonalidade em Time Series

Sazonalidade são ciclos que se repetem regularmente ao longo do tempo. Uma estrutura cíclica pode ou não ser sazonal, se consistentemente se repete na mesma frequência, então é, se não, é simplesmente chamado de um ciclo.

#### Benefícios para Machine Learning

Entender o componente sazonal pode melhorar a performance do modelo de duas formas principais:

- Sinal Limpo: Identificando e removendo o componente sazonal pode resultar em uma relação mais limpa entre variáveis de entrada e saída.
- Mais Informação: Mais informação sobre o componente sazonal pode provêr informação nova para melhorar a performance do modelo.

Modelando e removendo sazonalidade pode ser usada durante a fase de limpeza e preparação dos dados. Extração da sazonalidade pode criar novas input features, e pode ocorrer durante feature extraction ou feature engineering.

#### Tipos de Sazonalidade

Existem muitos tipos de sazonalidade, como por exemplo:

- Hora do Dia.
- Diariamente.
- Semanalmente.
- Mensalmente.
- Anualmente.

Identificar se tem ou não componente de sazonalidade é subjetivo. O modo mais simples de identificar é plotando e revisando os dados, em diferentes escalas e com ajuda das linhas de tendência.

#### Removendo Sazonalidade

Remover a sazonalidade é chamado de Ajuste Sazonal ou Dessazonalização. Para previsão e modelos preditivos, limita-se aos modelos que podem ser desenvolvidos em dados históricos e disponíveis quando são feitas previsões nos novos dados. Usaremos o **Minimum Daily Temperatures Dataset**.

#### Ajuste de Sazonalidade com Diferenciação

Um modo simples de corrigir um componente sazonal é por meio da diferenciação, tirando a sazonalidade de um dado de hoje ao subtrair com o dado da semana passada, por exemplo em casos de sazonalidade semanal. No caso do dataset, tempos sazonalidade a cada ano, mostrando balanço entre verão e inverno.

Podemos subtrair a temperatura mínima do mesmo dia do ano passado para corrigir a sazonalidade.

```python
# deseasonalize a time series using differencing
from pandas import read_csv
from matplotlib import pyplot
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,
                 parse_dates=True, squeeze=True)
X = series.values
diff = list()
days_in_year = 365
for i in range(days_in_year, len(X)):
    value = X[i] - X[i - days_in_year]
    diff.append(value)
pyplot.plot(diff)
pyplot.show()
```

#### Ajuste de Sazonalidade com Modelagem

O componente sazonal em uma certa série temporal provavelmente é uma onda senoidal sobre um período e amplitude fixa. Pode ser facilmente aproximado usando um curve-fitting method. Um dataset pode ser construído com o time index da onde senoidal como entrada, ou eixo X, e a observação como saída, ou eixo y. Por exemplo:

Time Index, Observation

1, obs1

2, obs2

3, obs3

4, obs4

5, obs5

Assim que está encaixado pelo método, o modelo pode ser usado para calcular o componente sazonal de qualquer índice de tempo. No caso de dados de temperatura, o índice de tempo seria o dia do ano. Podemos assim estimar o componente sazonal para o dia do ano de qualquer observação histórica ou futura. A curva pode ser usada como entrada para modelagem com algoritmos de supervised learning, ou subtraída das observações para criar séries ajustadas sazonalmente.

Vamos começar encaixando uma curva no dataset. A biblioteca NumPy tem uma função chamada **polyfit()** que pode ser usada para encaixar um polinomial de uma certa ordem para um dataset. Primeiro, podemos criar um dataset de índice temporal (dias nesse caso) para observação. Podemos pegar um ano ou todos os anos. Idealmente tentariamos os dois e veriamos qual o melhor.
Assim que o dataset está preparado, podemos criar o encaixe ao chamar **polyfit()** passando os valores do eixo X (integer dias do ano), valores do eixo Y (observações de temperatura), e a ordem do polinomial. A ordem controla o número de termos, e consequentemente a complexidade da curva. Para deixar simples será usado um polinomio de 4 ou 5 ordem. O modelo resultante tem a forma:

y = (x^4 X b1) + (x^3 X b2) + (x^2 X b3) + (x^1 X b4) + b5

Onde y é o valor de encaixe, x é o índice de tempo (dia do ano), e b1 até b5 são os coeficientes encontrados pelo algoritmo de otimização de encaixe de curva. Uma vez encaixado, teremos um conjunto de coeficientes que representam o modelo. Podemos usar esse modelo para calcular a curve de uma observação, um ano de observações ou o dataset inteiro. Exemplo completo listado abaixo:

```python
# model seasonality with a polynomial model
from pandas import read_csv
from matplotlib import pyplot
from numpy import polyfit
series = read_csv('daily-minimum-temperatures.csv', header=0, index_col=0,
                 parse_dates=True, squeeze=True)
# fit polynomial: x^2*b1 + x*b2 + ... + bn
X = [i%365 for i in range(0, len(series))]
y = series.values
degree = 4
coef = polyfit(X, y, degree)
print('Coefficients: %s' % coef)
# create curve
curve = list()
for i in range(len(X)):
    value = coef[-1]
    for d in range(degree):
        value += X[i]**(degree-d) * coef[d]
    curve.append(value)
# plot curve over original data
pyplot.plot(series.values)
pyplot.plot(curve, color='red', linewidth=3)
pyplot.show()
```

Rodando o exemplo foi criado o dataset, encaixado na curve, previsto os valores de cada dia no dataset, e depois plotado o modelo sazonal resultante sobre o dataset original.

Existem diversos métodos para verificar se uma série temporal é estacionária ou não:
- **Olhar os gráficos**: É possível verificar o gráfico da série temporal e checar visualmente se há alguma tendência ou sazonalidade óbvia.

#### Síntese Estatística

Um jeito rápido de ver se a série temporal é não estacionária é dividindo ela em dois ou mais partições e comparar a média e variância de cada grupo. Se elas diferem e a diferença é estatísticamente significante, a série temporal é não estacionária. Vamos tentar essa abordagem no dataset Daily Births.

#### Daily Births Dataset

Como estamos olhando para a média e variância, estamos assumindo que os dados conformam uma distribuição Gaussiana (também chamada de curva de bell ou distribuição normal). Podemos checar isso rápidamente olhando o histograma das observações.

```python
# plot a histogram of a time series
from pandas import read_csv
from matplotlib import pyplot
series = read_csv('daily-total-female-births.csv', header=0, index_col=0, parse_dates=True,
                 squeeze=True)
series.hist()
pyplot.plot()
```

Podemos ver claramente a distribuição Gaussiana. Em seguida dividimos a série temporal em duas sequências contíguas. Podemos então calcular a média e variância de cada grupo e comparar os valores.

```python
# calculate statistics of partitioned time series data
from pandas import read_csv
series = read_csv('daily-total-female-births.csv', header=0, index_col=0, parse_dates=True,
                 squeeze=True)
X = series.values
split = int(len(X) / 2)
X1, X2 = X[0:split], X[split:]
mean1, mean2 = X1.mean(), X2.mean()
var1, var2 = X1.var(), X2.var()
print('mean1=%f, mean2=%f' % (mean1, mean2))
print('variance1=%f, variance2=%f' % (var1, var2))
```

Podemos ver que os valores são diferentes, porém dentro de uma distância razoável. Vamos ver o mesmo com o dataset Airline Passengers

A seguir vamos ver um teste estatístico desenvolvido para explicitamente comentar se uma série temporal é estacionária.

#### Augmented Dickey-Fuller test

Testes estatísticos fazem fortes suposições sobre os dados. Eles só podem ser usados para informar o grau em que uma hipótese nula pode ser rejeitada. O resultado deve ser interpretado para ter significado. O teste Augmented Dickey-Fuller é um tipo de teste estatístico chamado teste de raíz unitária. A intuição por traz desse teste é que ele determina quão forte uma série temporal é definida por uma tendência. Ele usa um modelo autoregressivo e otimiza um critério de informação sobre múltiplos valores de lag diferentes. A hipótese nula é que a série temporal pode ser representada por uma raíz unitária, que é não estacionária (tem alguma estrutura dependente de tempo). A hipótese alternativa é que a série temporal é estacionária.

- **Hipótese Nula (H0)**: Falhando ao rejeitá-la, sugere que a série temporal tem uma raiz unitária, significando que é não estacionária. Tem alguma estrutura dependente de tempo.
- **Hipótese Alternativa (H1)**: A hipótese nula é rejeitada; sugere que a série temporal não tem uma raiz unitária, significando que ela é estacionária. Não tem uma estrutura dependente de tempo.

Interpretamos o resultado usando o p-value do teste. Um p-value abaixo de um limite (como 5% ou 1%)sugere que rejeitamos a hipótese nula (estacionária), de outro modo, um p-value acima do limite sugere que falhamos em rejeitar a hipótese nula (não estacionária).

- **p-value > 0.05**: Falhou ao rejeitar a hipótese nula (H0), os dados possuem uma raiz unitária e é não estacionária.
- **p-value <= 0.05**: Rejeita a hipótese nula (H0). os dados não possuem uma raiz unitária e é estacionária.

Abaixo um exemplo calculando o teste Augmented Dickey-Fuller no dataset Daily Female Births. A biblioteca Statsmodels provê uma função **adfuller()** que implementa o teste.

```python
# calculate stationarity test of time series data
from pandas import read_csv
from statsmodels.tsa.stattools import adfuller
series = read_csv('daily-total-female-births.csv', header=0, index_col=0, parse_dates=True,
                 squeeze=True)
X = series.values
result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

Quanto mais negativo o valor do teste, nesse caso -4, mais provável será a rejeição da hipótese nula (teremos um dataset estacionário). o valor ADF é menor que o valor crítico de 1%, isso também sugere essa rejeição.

## Modelos de Avaliação

### Capítulo 16: Backtest Forecast Models

Os métodos mais poderosos de machine learning não funcionam com dados de séries temporais, porque eles ignoram os componentes temporais inerentes desses problemas. Avaliar um modelo de machine learning em séries temporais é chamado de backtesting ou hindcasting.

#### Avaliação de um modelo

Ao avaliar um modelo para previsão de séries temporais, estamos interessados na performance dele nos dados que não foram usados para treiná-lo, isso é, os dados não vistos ou fora da amostra. Podemos fazer isso dividindo os dados em dois grupos, dando um grupo para o modelo treinar e fazendo ele prever o outro.

#### Monthly Sunspots Dataset

Nessa aula usaremos o dataset Monthly Sunspots, que descreve a contagem mensal do número de marcas observáveis no sol (sunspots) em um período de 230 anos (1749-1983). Colocar no diretório com o nome do arquivo de **sunspots.csv**.

#### Train-Test Split

É possível dividir o dataset em subsets de treino e de teste. O modelo pode ser preparado no dataset de treino e as previsões podem ser feitas e avaliadas no dataset de teste. Isso pode ser feito ao selecionar arbritáriamente um ponto de corte na lista ordenada de observações para criar os dois novos datasets.

Após carregar o dataset como uma Pandas **Series** , podemos extrair um array de dados NumPy. O ponto de corte pode ser calculado como um índice específico no array. Todos registros até o ponto de corte são pegos como o dataset de treino e todos os registros após o ponto de corte como o dataset de teste. Abaixo um exemplo usando um split de 66-34.

pagina 146

```python
#calculate a train-test split of a time series dataset
from pandas import read_csv
series = read_csv('sunspots.csv', header=0, index_col=0, parse_dates=True, squeeze=True)
X = series.values
train_size = int(len(X) * 0.66)
train, test = X[0:train_size], X[train_size:len(X)]
print('Observations: %d' % (len(X)))
print('Training Observations: %d' % (len(train)))
print('Testing Observations: %d' % (len(test)))
pyplot.plot(train)
pyplot.plot([None for i in train] + [x for x in test])
pyplot.show()
```

Esse método é rápido, preparação dos dados é simples e intuitivo, além de ter somente um modelo criado e avaliado. É bom quando há um grande volume de dados para que ambos os subsets sejam representativos.

#### Multiple Train-Test Splits

Podemos repetir o processo do train-test split múltiplas vezes para múltiplos modelos e assim colher uma estimativa mais robusta da performance esperada do método e configuração escolhidos. A biblioteca scikit-learn realiza esse processo de dividir o dataset em subsets em pontos de cortes diferentes de forma automática a partir do objeto **TimeSeriesSplit**. É preciso especificar o número de cortes a serem criados e o **TimeSeriesSplit** irá retornar os índices das observações treinadas e testadas de cada split requisitado.

O dataset sunspots tem 2.820 observações. Vamos criar 3 splits para o dataset. De acordo com a aritmética usada no objeto da scikit-learn, esperamos que os cortes a seguir sejam criados:

- **Split 1**: 705 train, 705 test
- **Split 2**: 1,410 train, 705 test
- **Split 3**: 2,115 train, 705 test

```python
# calculate repeated train-test splits of time series data
from pandas import read_csv
from sklearn.model_selection import TimeSeriesSplit
from matplotlib import pyplot
series = read_csv('sunspots.csv', header=0, index_col=0, parse_dates=True, squeeze=True)
X = series.values
splits = TimeSeriesSplit(n_splits=3)
pyplot.figure(1)
index = 1
for train_index, test_index in splits.split(X):
    train = X[train_index]
    test = X[test_index]
    print('Observations: %d' % (len(train) + len(test)))
    print('Training Observations: %d' % (len(train)))
    print('Testing Observations: %d' % (len(test)))
    pyplot.subplot(310 + index)
    pyplot.plot(train)
    pyplot.plot([None for i in train] + [x for x in test])
    index += 1
pyplot.show()
```

Usando esse método, mais modelos serão treinados, assim resultando em uma estimativa mais acurada da performance dos modelos em dados desconhecidos. Uma limitação desse modelo é que os modelos treinados permanecem fixos enquanto são avaliados em cada avaliação no conjunto de teste. Isso não é muito realístico, pois modelos podem ser retreinados enquanto novas observações diárias ou mensais se tornam disponíveis. Isso é o que será abordado na próxima seção.

#### Walk Forward Validation

Na prática, muito provavelmente iremos retreinar nosso modelo assim que novos dados se tornam disponíveis. Isso dá uma melhor chance do modelo fazer boas previsões em cada passo de tempo. Podemos avaliar nosso modelo de machine learning sobre essa suposição. Há algumas decisões a tomar:

- **Minimum Number of Observations**. Primeiro, devemos selecionar o número mínimo de observações necessárias para treinar o modelo. É como se fosse a largura de janela se uma janela deslizante fosse usada.
- **Janela Deslizante ou Expansiva**. Em seguida, precisamos decidir se o modelo vai ser treinado com todos os dados disponíveis ou só com os mais recentes. Isso determina se uma janela deslizante ou expansiva será usada.

Depois de escolher as configurações, modelos podem ser treinados e avaliados.

1. Começando do início da time series, o número mínimo de amostras na janela é usado para treinar o modelo.
2. O modelo faz a previsão do próximo passo de tempo.
3. A previsão é armazenada ou avaliada contra o valor conhecido.
4. A janela é expandida para incluir o valor conhecido e o processo é repetido (ir para passo 1).

Como essa metodologia envolve se mover ao longo da time series um passo de cada vez, é chamada de Walk Forward Testing ou Walk Forward Validation. Adicionalmente, por utilizar uma janela deslizante ou expansiva para treinar o modelo, esse método também é referido como Rolling Window Analysis ou Rolling Forecast. Essa habilidade não está disponível no scikit-learn, porém pode chegar no mesmo efeito se cautelosamente configurar uma TimeSeriesSplit. Abaixo um exemplo de como dividir os dados em conjuntos de treino e teste usando o método Walk Forward Validation.

```python
# walk forward evaluation model for time series data
from pandas import read_csv
series = read_csv('sunspots.csv', header=0, index_col=0, parse_dates=True, squeeze=True)
X = series.values
n_train = 500
n_records = len(X)
for i in range(n_train, n_records):
    train, test = X[0:i], X[i:i+1]
    print('train=%d, test=%d' % (len(train), len(test)))
```

O exemplo imprime os valores de train e test, foi criado 2.820 modelos, isso é um modo mais robusto de avaliar como eles irão se comportar na prática. Tem um alto valor computacional, então tem que tomar cuidado com o tamanho da janela e o seu tipo. Esse é o padrão de ouro da avaliação de modelo, é recomendado para projetos próprios.

### Capítulo 17: Medições de Desempenho da Previsão

Medições de performance da previsão de séries temporais resumem a habilidade e capacidade do modelo de previsão que faz as previsões. Séries temporais geralmente focam na previsão de valores reais, chamados problemas de regressão. Portanto as medições de performance nesse tutorial irão focar nos métodos para avaliar previsões de valores reais.

#### Forecast Error (ou Residual Forecast Error)

O Forecast error é calculado como o valor esperado menos o valor previsto. Isso é chamado de erro residual da previsão.

forecast_error = expected_value - predicted_value

O forecast error pode ser calculado para cada previsão, provendo uma série temporal de forecast errors. O exemplo abaixo demonstra como o frecast error pode ser calculado para uma série de 5 previsões comparadas com 5 valores esperados.

```python
# calculate forecast error
expected = [0.0, 0.5, 0.0, 0.5, 0.0]
predictions = [0.2, 0.4, 0.1, 0.6, 0.2]
forecast_errors = [expected[i]-predictions[i] for i in range(len(expected))]
print('Forecast Errors: %s' % forecast_errors)
```

#### Mean Forecast Error (ou Forecast Bias)

Mean forecast error é calculado como a média dos valores de forecast error.

mean_forecast_error = mean(forecast_error)

Forecast errors podem ser positivos ou negativos. Isso siginifica que quando a média é calculada um mean forecast error ideal seria zero. Um valor de média de forecast error diferente de zero sugere uma tendência do modelo de sobre prever (error negativo) ou sub prever (error positivo). Por isso é chamado também de forecast bias (viés da previsão). Abaixo exemplo de como calcular isso manualmente.

```python
# calculate mean forecast error
expected = [0.0, 0.5, 0.0, 0.5, 0.0]
predictions = [0.2, 0.4, 0.1, 0.6, 0.2]
forecast_errors = [expected[i]-predictions[i] for i in range(len(expected))]
bias = sum(forecast_errors) * 1.0/len(expected)
print('Bias: %f' % bias)
```

Resultado negativo quer dizer que há uma sobre previsão. Quando é perto do zero quer dizer que não tem viés.

#### Mean Absolute Error

O mean absolute error, ou MAE, é calculado como a média dos valores de erro da previsão, onde todos os valores de previsão são forçados a serem positivos. Forçando eles a serem positivos é chamado de fazer eles absolutos. Isso é significado com a função do absoluto **abs()** ou mostrado matematicamente como dois caracteres pipe em volta do valor: **|valor|**.

mean absolute error = mean(abs(forecast error))

Onde **abs()** torna o valor positivo, **forecast_error** é um ou uma sequência de erros de previsão, e **mean()** calcula o valor da média. Podemos usar a função **mean_absolute_error()** da biblioteca scikit-learn para calcular a mean absolute error para uma lista de previsões. O exemplo abaixo demonstra essa função.

```python
# calculate mean absolute error
from sklearn.metrics import mean_absolute_error
expected = [0.0, 0.5, 0.0, 0.5, 0.0]
predictions = [0.2, 0.4, 0.1, 0.6, 0.2]
mae = mean_absolute_error(expected, predictions)
print('MAE: %f' % mae)
```

Esse valor está na mesma unidade dos valores previstos. Uma mean absolute error de zero indica que não houve erros.

#### Mean Squared Error

O mean squared error, ou MSE, é calculado como a média do quadrado dos valores de erro da previsão. Fazendo o quadrado dos valores força-os a se tornarem positivos. Também tem o efeito de pôr mais peso em erros maiores. Erros muito grandes ou fora da curva são dobrados, que por sua vez tem o efeito de puxar a média do quadrado dos erros de previsão para fora, resultando numa pontuação maior de mean squared error. Em prática, a pontuação dá performance pior para aqueles modelos que fazem previsões mais erradas.

mean squared error = mean(forecast error^2)

Podemos usar a função **mean_squared_error()** da biblioteca scikit-learn para calcular a média quadrada dos erros de uma lista de previsões. Abaixo um exemplo:

```python
# calculate mean squared error
from sklearn.metrics import mean_squared_error
expected = [0.0, 0.5, 0.0, 0.5, 0.0]
predictions = [0.2, 0.4, 0.1, 0.6, 0.2]
mse = mean_squared_error(expected, predictions)
print('MSE: %f' % mse)
```

MSE em zero significa uma pontuação perfeita e nenhum erro.

#### Root Mean Squared Error

O mean squared error descrito ascima está nas unidades quadradas das previsões. Pode ser transformada de volta à unidade original das previsões ao tirar a raiz quadrada da pontuação do mean squared error. Isso é chamado de root mean squared error, ou RMSE.

Isso pode ser calculado usando a função matemática **sqrt()** no mean squared error calculado usando a função **mean_squared_error()** da scikit-learn.

```python
# calculate root mean squared error
from sklearn.metrics import mean_squared_error
from math import sqrt
expected = [0.0, 0.5, 0.0, 0.5, 0.0]
predictions = [0.2, 0.4, 0.1, 0.6, 0.2]
mse = mean_squared_error(expected, predictions)
rmse = sqrt(mse)
print('RMSE: %f' % rmse)
```

### Capítulo 18: Persistence Model para Previsão

É importante estabelecer uma baseline em qualquer problema de previsão de time series. Uma baseline em performance te dá uma ideia de quão bem todos os outros modelos irão performar no seu problema. Nesse tutorial vamos ver como desenvolver uma persistence forecast que pode usar para calcular um nível baseline de performance em um dataset de time series com Python.

#### Forecast Performance Baseline

Uma baseline em performance de previsão provê um ponto de comparação. Se um modelo alcança performance na baseline ou abaixo, ele deve ser modificado ou abandonado. A técnica usada para gerar uma previsão para calcular a baseline deve ser fácil de implementar e genérica. Antes de desenvolver uma baseline é preciso desenvolver um equipamento de testes. Ele é composto de:

1. O dataset que você pretende usar para treinar e avaliar os modelos.
2. A técnica de resampling que pretende-se usar para estimar a performance da técnica (ex. train/test split).
3. A medida de performance que pretende usar para avaliar as previsões (ex. root mean squared error).

Uma vez preparado, você precisa então selecionar uma técnica *naive* que você pode usar para fazer uma previsão e calcular a baseline de performance. O objetivo é obter uma baseline no seu problema de previsão de time series o mais rápido possível para que possa voltar ao trabalho de melhor entender o dataset e desenvoler modelos mais avançados. Três propriedades de uma boa técnica para fazer uma baseline de performance são:

- **Simples**: Um método que requer pouco ou nenhum treinamento ou inteligência.
- **Rápido**: Um método que é rápido de ser implementado e computacionalmente trivial para se fazer uma previsão.
- **Repetitível**: Um método que é determinístico, significando que produz uma saída esperada dado uma mesma entrada.

Um algoritmo comum usad para estabelecer uma baseline de performance é o persistence algorithm.

#### Persistence Algorithm

O método mais comum de baseline para supervised machine learning é o algoritmo Zero Rule. Esse algoritmo prevê a classe majoritária no caso de classificação, ou a saída média no caso de regressão. Isso poderia ser usado em time series, mas ele não respeita a estrutura de correlação temporal de um dataset desses. A técnica equivalente para se usar com time series é o persistence algorithm.

O persistence algorithm usa o valor do passo de tempo atual (**t**) para prever a saída esperada no próximo passo de tempo (**t+1**). Isso satisfaz as três condições acima para uma baseline de previsão. Para fazer isso concreto, vamos olhar em como desenvolver um persistence model e usar ele para estabelecer uma baseline de performance para um problema univariado simples de time series.
